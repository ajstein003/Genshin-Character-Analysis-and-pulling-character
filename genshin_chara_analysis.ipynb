```python
import matplotlib.pyplot as plt
import scipy
import seaborn as sn
import warnings
warnings.filterwarnings(action='ignore')

genshin=pd.read_csv('/content/Genshin_Impact_All_Character_Stat.csv')
```

```python
genshin.info()
```

```python
lv90_base=genshin.loc[(genshin.Lv==90)&(genshin.Rarity==4),:]
lv90_base
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sn.barplot(x='Character',y='Base HP',data=lv90_base.sort_values('Base HP'))
plt.title('Level - 90 Base HP by Character')
plt.xticks(rotation=70,fontsize=13)
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sn.barplot(x='Character',y='Base ATK',data=lv90_base.sort_values('Base ATK'))
plt.title('Level - 90 Base ATK by Character')
plt.xticks(rotation=70,fontsize=13)
```

```python
fig,ax=plt.subplots(figsize=(10,8))
sn.barplot(y='Character',x='Base DEF',data=lv90_base.sort_values('Base DEF'),orient='h',)
sn.set_theme(style='whitegrid')
plt.title('Level - 90 Base DEF by Character')
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,8))
lv90_base.loc[:,'Element'].value_counts().plot(kind='bar')
plt.xticks(rotation=50,fontsize=13)
plt.title('Element Type of 4-Rarity')
plt.xlabel('Elenment Type')
plt.ylabel('Frequency')
```
